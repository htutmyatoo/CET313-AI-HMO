## Exercise 2: Write a step-by-step instruction to find the min number of coins. Write your solution in Jupyter notebook as a solution to exercise 2.

**Given denominations:** 1p, 2p, 5p, 10p, 20p, 50p, 100p, 200p

**Find:** the minimum number of coins to make a 93p = ?

**Step-by-step instruction for my Solution:**
- Start with the largest denomination possible that is less than the target amount. In this case, the target amount is 93p. The largest denomination less than 93p is 50p.
- Use the largest denomination coin as many times as possible without exceeding the target amount. We can use one 50p coin (93p - 50p = 43p).
- Move on to the next lower denomination that is still greater than the remaining amount. We have 43p remaining, so consider the 20p coin.
- Use the next lower denomination coin as many times as possible without exceeding the remaining amount. We can use two 20p coins (43p + 40p = 83p).
- Continue iterating through the remaining denominations one at a time until the exact target amount is reached. We need 1 x 2p coin (83p + 2p = 85p) and 1 x 1p coin (85p + 1p = 86p = 93p).
    
By following these steps, we can achieve the minimum number of coins needed (93p), which is 5:
- 1 x 50p coin
- 2 x 20p coins
- 1 x 2p coin
- 1 x 1p coin

## Exercise 3: Use the following python script to solve the problem.

````python
def coin_change(coins, amount):
  coins.sort(reverse=True) # Sort the coins in descending order
  coin_count = 0 # Variable to keep track of the total number of coins used
  change = [] # List to store the coins used

  for coin in coins:
    while amount >= coin:
      amount -= coin
      coin_count += 1
      change.append(coin)

  if amount == 0:
    print(f"Minimum number of coins required: {coin_count}")
    print("Coins used:", change)
  else:
    print("Not possible to get the desired change with the coins.")

# Example usage
coins = [1, 2, 5, 10, 20, 50, 100, 200] # List of available coins
amount = 93 # Amount for which we need to find the minimum number of coins

coin_change(coins, amount)
````

The previous script tackles the coin change problem using a ```greedy approach```. It first sorts the available denominations ```(coins)``` in descending order. Then, it iterates through each coin, subtracting its value from the target amount ```(amount)``` as long as it fits. ## Exercise 6: Is there anything you would change in to enhance it.

Yes, I found a few ways to potentially enhance the script:
- **Adding Early Termination** - to improve the performance by avoiding unnecessary iterations once the target amount is reached.
- **Tracking Remainder** - to track the remaining amount for clarity
- **Adding Coin Count Dictionary ```coin_counter```** - to count the number of each coin used to provide a clearer and more detailed output.
- **Making Output Detail** - to understand easier
- **Enhancing Comments and Variable Names** - for better readability and maintainability.

The enhanced script:

````python
def coin_change(coins, amount):
    # Sort the coins in descending order
    coins.sort(reverse=True)
    coin_count = 0  # Variable to keep track of the total number of coins used
    change = []  # List to store the coins used
    remainder = amount  # Track remaining amount
    
    # Dictionary to count each denomination used
    coin_counter = {coin: 0 for coin in coins}

    for coin in coins:
        while remainder >= coin:
            remainder -= coin
            coin_count += 1
            coin_counter[coin] += 1
            change.append(coin)  # Add coin to the list of coins used

        # Early termination if remainder is 0
        if remainder == 0:
            break

    if remainder == 0:
        print(f"Minimum number of coins required: {coin_count}")
        print("Coins used:", change)
        for coin, count in coin_counter.items():
            if count > 0:
                print(f"{count} x {coin}p")
    else:
        print("Not possible to get the desired change with the coins.")

coins = [1, 2, 5, 10, 20, 50, 100, 200]  # List of available coins
amount = 95  # Amount for which the need to find the minimum number of coins
coin_change(coins, amount)

print()

coins = [1, 2, 5, 10, 20, 50, 100, 200]  # List of available coins
amount = 150  # Amount for which the need to find the minimum number of coins
coin_change(coins, amount)

print()

coins = [1, 2, 5, 10, 20, 50, 100, 200]  # List of available coins
amount = 255  # Amount for which the need to find the minimum number of coins
coin_change(coins, amount)
````
