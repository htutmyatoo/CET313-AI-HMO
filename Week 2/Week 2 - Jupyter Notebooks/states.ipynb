```python
import json
from kanren import run, fact, eq, Relation, var, conde

adjacent = Relation()
coastal = Relation()

file_coastal = 'coastal_states.txt'
file_adjacent = 'adjacent_states.txt'

# Read the file containing the coastal states 
with open(file_coastal, 'r') as f:
    line = f.read()
    coastal_states = line.split(',')

# Add the info to the fact base
for state in coastal_states: 
    fact(coastal, state)

# Read the file containing the coastal states
with open(file_adjacent, 'r') as f:
    adjlist = [line.strip().split(',') for line in f if line and line[0].isalpha()]

# Add the info to the fact base
for L in adjlist:
    head, tail = L[0], L[1:]
    for state in tail:
        fact(adjacent, head, state)

# Initialize the variables
x, y = var(), var()
```

#### Example 1: Is Nevada adjacent to Louisiana?

```python
output = run(0, x, adjacent('Nevada', 'Louisiana')) 
print("Is Nevada adjacent to Louisiana?:") 
print('Yes' if len(output) else 'No')
```

#### Example 2: States adjacent to Oregon

```python
output = run(0, x, adjacent('Oregon', x)) 
print('List of states adjacent to Oregon:') 
for item in output:
    print(item)
```

#### Example 3: States adjacent to Mississippi that are coastal

```python
output = run(0, x, adjacent('Mississippi', x), coastal(x)) 
print('List of coastal states adjacent to Mississippi:') 
for item in output:
    print(item)
```

#### Example 4: List of 'n' states that border a coastal state

```python
n = 7
output = run(n, x, coastal(y), adjacent(x, y))
print('List of ' + str(n) + ' states that border a coastal state:') 
for item in output:
    print(item)
```

#### Example 8: Find states present in both files

```python
coastal_states = set()  # Create an empty set to store coastal states
with open('coastal_states.txt', 'r') as f:
  coastal_states.update(f.read().split(','))  # Add states to the set

all_states = set()  # Create an empty set to store all states
for line in open('adjacent_states.txt'):
  if line and line[0].isalpha():  # Skip empty lines and lines not starting with a letter
    all_states.update(line.strip().split(','))  # Add states to the set

common_states = coastal_states.intersection(all_states)  # Find intersection

print('States present in both coastal_states.txt and adjacent_states.txt:')
for state in common_states:
  print(state)
```

#### Example 9: Check if the user-provided state is coastal

```python
# Get user input for state
user_state = input("\Start.." + "\nEnter a state name (or 'q' to quit): ")

while user_state != 'q':
    
  output = run(0, user_state, coastal(user_state))

  if output:
    print(user_state + " is a coastal state.")
  else:
    print(user_state + " is not a coastal state.")

  # Get user input for the next state or quit
  user_state = input("Enter another state name (or 'q' to quit): ")

print("--- End ---")
```
