## Exercise 15: Using the previous examples develop a text summarisation script that would read a text file and summarise it into 3 sentences.

```python
# Read or Get a text file
def read_text_file(file_path):
    try:
        with open(file_path, 'r') as file:
            text = file.read()
            return text
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")

file_path = "hacker_culture.txt"
before_summarize_text = read_text_file(file_path)
```

### A. Tokenize the sentences and words in the text.

```python
# Step A: Tokenize the sentences and words in the text.
def tokenize_text(text):
    sentences = sent_tokenize(text)
    words = [word_tokenize(sentence) for sentence in sentences]
    return sentences, words

# Tokenize the sentences and words in the text.
sentences, words = tokenize_text(before_summarize_text)
print("Tokenized Sentences:")
print(sentences)
print("\nTokenized Sentences:")
print(words)
```

### B. Removes any stopwords from the text

```python
def remove_stopwords(words):
    stop_words = set(stopwords.words('english'))
    filtered_words = [[word for word in word_list if word.lower() not in stop_words] for word_list in words]
    return filtered_words

# Remove any stopwords from the text.
filtered_words = remove_stopwords(words)
print(f"Filtered Words (Stopwords Removed): \n\n{filtered_words}")
```

### C. Calculates the frequencies of each word, excluding the stopwords.

```python
from collections import defaultdict, Counter

def calculate_word_frequencies(filtered_words):
    word_freq = Counter(word for word_list in filtered_words for word in word_list)
    return word_freq

# Calculate the frequencies of each word, excluding the stopwords.
word_freq = calculate_word_frequencies(filtered_words)
for word, count in word_freq.items():
  print(f"{word}: {count}")
```

### D. Calculates the score for each sentence based on the frequencies of the words it contains, considering only sentences with less than 30 words.

```python
def calculate_sentence_scores(sentences, filtered_words, word_freq):
    sentence_scores = defaultdict(float)
    for i, word_list in enumerate(filtered_words):
        if len(word_list) < 30:
            for word in word_list:
                sentence_scores[sentences[i]] += word_freq[word]
    return sentence_scores

# Calculate the score for each sentence based on the frequencies of the words it contains,
# considering only sentences with less than 30 words.
sentence_scores = calculate_sentence_scores(sentences, filtered_words, word_freq)
print(f"\nSentence Scores: \n\n{sentence_scores}")
```

### E. Selects the top 2 sentences with the highest scores to generate a summary.

```python
def get_top_sentences(sentence_scores, top_n):
    top_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)[:top_n]
    return top_sentences

# Select the top 2 sentences with the highest scores to generate a summary.
top_sentences = get_top_sentences(sentence_scores, 2)
print(f"\nTop Sentences: \n{top_sentences}")
```

### F. Combines these top sentences to create the final summary.

```python
def create_summary(top_sentences):
    return ' '.join(top_sentences)

# Combine these top sentences to create the final summary.
summary = create_summary(top_sentences)
print(f"Summary: {summary}")
```

### A text summarisation script that would read a text file and summarise it into 3 sentences.

```python
# Main function to summarize the text
def summarize_text(text):
    sentences, words = tokenize_text(text)
    filtered_words = remove_stopwords(words)
    word_freq = calculate_word_frequencies(filtered_words)
    sentence_scores = calculate_sentence_scores(sentences, filtered_words, word_freq)
    top_sentences = get_top_sentences(sentence_scores, 3)
    summary = create_summary(top_sentences)
    return summary

# Read the text file and summarize the content
file_path = "hacker_culture.txt"
before_summarize_text = read_text_file(file_path)
if before_summarize_text:
    summary = summarize_text(before_summarize_text)
    print(f"Summary: {summary}")
    
```
