## Exercise 3: Initially we will start with a simple k-means algorithm implantation.

````python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs 
from sklearn.cluster import KMeans 

random_state_value = 42 
# Generate random data for clustering 
np.random.seed(random_state_value) 
data, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, 
                     random_state=42) 

# Function to perform K-means clustering and plot the results 
def plot_kmeans_clusters(data, n_clusters): 
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state_value) 
    labels = kmeans.fit_predict(data) 
    
    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', 
                edgecolor='k', s=50) 
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
                c='red', marker='x', s=200, label='Centroids') 
    plt.title(f'K-means Clustering with {n_clusters} Clusters') 
    plt.legend() 
    plt.show() 
    
plot_kmeans_clusters(data, n_clusters=2)
````

## Exercise 5: Update the script to cluster the data with different cluster number ranging from 2 clusters to 10 (using loop).

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

random_state_value = 42

# Generate random data for clustering
np.random.seed(random_state_value)
data, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0,
                    random_state=42)


def plot_kmeans_clusters(data):
    for n_clusters in range(2, 11):  # Loop from 2 to 10 clusters
        kmeans = KMeans(n_clusters=n_clusters, random_state=random_state_value)
        labels = kmeans.fit_predict(data)

        plt.figure()  # Create a new figure for each cluster number
        plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis',
                    edgecolor='k', s=50)
        plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
                    c='red', marker='x', s=200, label='Centroids')
        plt.title(f'K-means Clustering with {n_clusters} Clusters')
        plt.legend()
        plt.show()


plot_kmeans_clusters(data)
````

## Exercise 6: Experiment with different random state value and comment on its effect on the algorithm.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

def plot_kmeans_clusters(data, n_clusters, random_state_value):

    np.random.seed(random_state_value)
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state_value)
    labels = kmeans.fit_predict(data)

    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis',
                edgecolor='k', s=50)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
                c='red', marker='x', s=200, label='Centroids')
    plt.title(f'K-means Clustering with {n_clusters} Clusters (Random State: {random_state_value})')
    plt.legend()
    plt.show()

# Generate random data for clustering
n_samples = 300
centers = 3
cluster_std = 1.0

# Run K-means clustering with different random states
random_states = [42, 100, 200]
for random_state_value in random_states:
    data, _ = make_blobs(n_samples=n_samples, centers=centers, cluster_std=cluster_std, random_state=random_state_value)
    plot_kmeans_clusters(data, n_clusters=2, random_state_value=random_state_value)

plt.show()  # Ensure final plot is displayed
````

### Data Generation

````python
import numpy as np

from sklearn.datasets import make_blobs

n_samples = 1500
random_state = 170
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]

X, y = make_blobs(n_samples=n_samples, random_state=random_state)
X_aniso = np.dot(X, transformation)  # Anisotropic blobs
X_varied, y_varied = make_blobs(
    n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=random_state
)  # Unequal variance
X_filtered = np.vstack(
    (X[y == 0][:500], X[y == 1][:100], X[y == 2][:10])
)  # Unevenly sized blobs
y_filtered = [0] * 500 + [1] * 100 + [2] * 10
````

````python
# Visualization
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

axs[0, 0].scatter(X[:, 0], X[:, 1], c=y)
axs[0, 0].set_title("Mixture of Gaussian Blobs")

axs[0, 1].scatter(X_aniso[:, 0], X_aniso[:, 1], c=y)
axs[0, 1].set_title("Anisotropically Distributed Blobs")

axs[1, 0].scatter(X_varied[:, 0], X_varied[:, 1], c=y_varied)
axs[1, 0].set_title("Unequal Variance")

axs[1, 1].scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_filtered)
axs[1, 1].set_title("Unevenly Sized Blobs")

plt.suptitle("Ground truth clusters").set_y(0.95)
plt.show()
````

### Fit models and plot results

````python
from sklearn.cluster import KMeans

common_params = {
    "n_init": "auto",
    "random_state": random_state,
}

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

y_pred = KMeans(n_clusters=2, **common_params).fit_predict(X)
axs[0, 0].scatter(X[:, 0], X[:, 1], c=y_pred)
axs[0, 0].set_title("Non-optimal Number of Clusters")

y_pred = KMeans(n_clusters=3, **common_params).fit_predict(X_aniso)
axs[0, 1].scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
axs[0, 1].set_title("Anisotropically Distributed Blobs")

y_pred = KMeans(n_clusters=3, **common_params).fit_predict(X_varied)
axs[1, 0].scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
axs[1, 0].set_title("Unequal Variance")

y_pred = KMeans(n_clusters=3, **common_params).fit_predict(X_filtered)
axs[1, 1].scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred)
axs[1, 1].set_title("Unevenly Sized Blobs")

plt.suptitle("Unexpected KMeans clusters").set_y(0.95)
plt.show()
````

### Possible solutions

````python
y_pred = KMeans(n_clusters=3, **common_params).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Optimal Number of Clusters")
plt.show()
````

````python
y_pred = KMeans(n_clusters=3, n_init=10, random_state=random_state).fit_predict(
    X_filtered
)
plt.scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred)
plt.title("Unevenly Sized Blobs \nwith several initializations")
plt.show()
````

````python
from sklearn.mixture import GaussianMixture

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

y_pred = GaussianMixture(n_components=3).fit_predict(X_aniso)
ax1.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
ax1.set_title("Anisotropically Distributed Blobs")

y_pred = GaussianMixture(n_components=3).fit_predict(X_varied)
ax2.scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
ax2.set_title("Unequal Variance")

plt.suptitle("Gaussian mixture clusters").set_y(0.95)
plt.show()
````

### Key Points

#### K-means Assumptions:
- K-means assumes that the data clusters are roughly spherical (round) or have a similar shape in all directions.
- It also assumes equal variance within clusters, meaning the data points within a cluster are spread out similarly around the centroid.

#### Data Generation:
- The tutorial first generates sample data using ```make_blobs``` from scikit-learn. This function creates spherical clusters by default.
- It then applies transformations or manipulations to create datasets that violate the assumptions:
    - Anisotropic blobs: The data is stretched or distorted, creating elongated clusters.
    - Unequal variance: Clusters have different spreads, with some being tighter and others more spread out.
    - Unevenly sized blobs: Clusters have significantly different numbers of data points.

#### Visualization of Impact:
- The tutorial uses K-means with different numbers of clusters (```n_clusters```) on both the original and manipulated datasets.
- It then visualizes the resulting clusters using scatter plots. This allows us to see how K-means struggles to identify appropriate centroids and separate clusters when the assumptions are violated.
- The visualizations clearly show how K-means can produce:
    - Misplaced centroids in elongated clusters (anisotropic blobs).
    - Inaccurate boundaries and potential overlap between clusters with unequal variance.
    - Uneven cluster assignments due to imbalanced cluster sizes.

#### Alternative Clustering Methods:
- The tutorial doesn't explicitly mention alternatives, but it highlights the limitations of K-means when these assumptions are violated.
- Other clustering algorithms, like DBSCAN (Density-Based Spatial Clustering of Applications with Noise), might be better suited for data with non-spherical shapes or uneven densities.

In summary, this tutorial serves as a cautionary tale for relying solely on K-means clustering. It emphasizes the importance of understanding the underlying assumptions and choosing an appropriate algorithm based on our data's characteristics.
